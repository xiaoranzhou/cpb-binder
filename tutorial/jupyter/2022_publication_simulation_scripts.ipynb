```python
# Step 2

def total_sim(name, time_total, resistance_parameter, loading_vml, unloading_vmu):
    # stage 1
    plant = pb.Plant()
    plant.openXML("../../modelparameter/plant/anatomy-growth/" + name)
    seeds = plant.getOrganRandomParameter(pb.OrganTypes.seed)
    roots = plant.getOrganRandomParameter(pb.OrganTypes.root)
    stems = plant.getOrganRandomParameter(pb.OrganTypes.stem)
    leafs = plant.getOrganRandomParameter(pb.OrganTypes.leaf)
    plant.initialize(True)

    subtype_to_growth_rate = [0]*10
    for i,root_para in enumerate(roots) :
        subtype_to_growth_rate[root_para.subType] = root_para.r
        root_para.r = 0.25 # initial growthrate is limited.
    

   


    column_name = ["time", 
                   "id", 
                   "organ_type", 
                   "sub_type", 
                   "connect_no", 
                   "organ_id",
                   "organ_id2",
                   "parent_organ",
                   "length", 
                   "res_main", 
                   "carbon_q",
                   "c_satis",

                   "rw_satis",
                   "c_demand_2" ,                          
                   "growth_rate", 
                   "exudation", 
                   "order",
                   "organ_length",
                   "coor_x", 
                   "coor_y", 
                   "coor_z", 
                   "connection_1", 
                   "connection_2",
                   "radius",

                   "r_xyl_node",
                   "radial_water",
                   "r_xyl",
                   "r_st",

                  ]
    pd_output =  pd.DataFrame( [], columns= column_name )

    extent = 5000 # parameter to increase the plant size
    r_trsv = 100 # parameter of resistance between xylem and phloem (apoplasmic pathway )
    day_i = 0 # starting day
    vml = 6.40e-06 *extent*6 # maximal loading rate  Michaelis-Menten-Paremter
    vmu =  1e+98 # maximal unloading rate  Michaelis-Menten-Paremter
    vol_st = 2.6e-05 *5000 # sieve tube volume
    r_phl_mb = 0.028 *5 # cross-membrane pathway from phloem apoplasm to sieve tube 
    g_propotion = lambda water_potential : (20.79 + 83.597* 2.71828**(1.39578* (water_potential)) )/100

    for time in range(1, time_total): # i*days to simulation
        pd_output_1day =  pd.DataFrame( [], columns= column_name )

        day_i = day_i +1
        day = 1# the step of every loop (day)
        hour = 24 # the hours of every loop (hour) used in PiafMunch
        plant.simulate(1 * day) # start simulation
        organ = plant.getOrgans()
        ana = pb.SegmentAnalyser(plant)
        ana.write("{}_{}_{}_{}_{}.vtp".format(name, time_total, resistance_parameter, loading_vml, unloading_vmu)) #writes vtp the name is followed by the number of days simulated


        dict_all  = convert( plant ) # convert plant to python outputs
        nodes_organtype = dict_all['nodes_organtype']
        nodes_coor = python_nodes(plant)
        segs = np.array([np.array(s) for s in plant.getSegments()])
        node_connection_o = segs
        pd_output_1day.loc[:,"id"] = list(map(int,dict_all["nodes_organtype"].T[0]))
        pd_output_1day.loc[:,"organ_type"] = list(map(int,dict_all["nodes_organtype"].T[1]))
        pd_output_1day.loc[:,"connect_no"] = list(map(int,dict_all["nodes_organtype"].T[2]))

        pd_output_1day.loc[:,"coor_x"] = list(nodes_coor.T[3])
        pd_output_1day.loc[:,"coor_y"] = list(nodes_coor.T[4])
        pd_output_1day.loc[:,"coor_z"] = list(nodes_coor.T[5])

        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"connection_1"] = list(map(int,dict_all["node_connection"].T[0]))
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"connection_2"] = list(map(int,dict_all["node_connection"].T[1]))

        piafmunch_outputname = name+str(day_i) +".ini" # set the name of the parameters
        N1R_node = nodes_organtype[(nodes_organtype[:,1] ==2 ) & (nodes_organtype[:,3] ==1 )] # remove seed. get roots with one connection
        
        Soil_water = [0.1]*(len(dict_all['node_connection'])*2+ len(N1R_node) +1)
        organ_radius = [0]*((len(organ))) # initiate the radius of each segment
        k1 = [0]*(len(dict_all['node_connection'])+1) # k1 parameter in equation
        kt = [0]*(len(dict_all['node_connection'])+1) # total carbon maintenance

        node_id_to_organ = [0]*plant.getNumberOfNodes() #
        node_id_to_subtype = [0]*plant.getNumberOfNodes() #
        node_id_to_organ2 = [0]*plant.getNumberOfNodes() #



        ########################################## python nodes conversion #########################################
        nodes_length = np.zeros(len(nodes_coor))
        for i in range(1,len(nodes_coor)):
            if nodes_coor[i][1]==2: #if the organtype is root
                nodes_length[i] = (nodes_length[node_connection_o[(node_connection_o[:,1] == i)][0][0]]  
                                   + ((nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][3] -nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][3])**2 
                                    + (nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][4] -nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][4])**2 
                                    + (nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][5] -nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][5])**2)**(0.5) )
            else:
                nodes_length[i] = (nodes_length[node_connection_o[(node_connection_o[:,1] == i)][0][0]] 
                                   + ((nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][3]-nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][3])**2 
                                      + (nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][4]-nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][4])**2 
                                      + (nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][0]][5]-nodes_coor[node_connection_o[(node_connection_o[:,1] == i)][0][1]][5])**2)**(0.5) )
        N1L_node = nodes_organtype[ ((nodes_organtype[:,1] ==4)|(nodes_organtype[:,1] ==3)  ) & (nodes_organtype[:,3] ==1 )] # remove seed. get stem or leaf with one connection


########################## nodewise iteration to know which organ it belongs to ###########################################
########################## nodewise iteration to get maximal  ###########################################  

        for k in reversed(range(len(organ))):  # loop through all the organs
            organ_radius[k] =  0.05 #organ[k].getParam().a #the radius of one organ
            j=0

            for j in range(1,organ[k].getNumberOfNodes()): # loop through all segments in one organs
                node_id = organ[k].getNodeId(j)
                node_id_to_organ[node_id]=k # set an id, so we can get the organ number from a node number
                node_id_to_organ2[node_id]=organ[k].getId()
                node_id_to_subtype[node_id] = int(organ[k].getParameter("subType"))
                connection_belong = np.where(node_connection_o.T[1][:] == node_id )
                first_node = node_connection_o[connection_belong[0][0]][0]
                node_length= nodes_length[node_id] - nodes_length[first_node]
                
                
                volume = organ_radius[k]**2*3.1415926* abs(node_length)

                m_r = 0.0032 # Organ maintenance respiration rate (gCO2 g1 DM)
                mass_per_volume = 0.10 #root dry mass per unit volume (gDM cm3) 0.05-0.14
                Carbon_main = volume*0.1 *0.0032/(24) *1000/44/6  # (1000/12) from grams/day/cm3 co2 to mmol c/hour/cm3 sucrose calculate the maintenance by volume (mmol)
                k1[organ[k].getNodeId(j)] = 0
############################# water potential in left and right !root! ##################################################

        dry_soil = resistance_parameter
        wet_soil = unloading_vmu

        

    #     here need more development 
        N1R_node = nodes_organtype[(nodes_organtype[:,1] ==2 ) & (nodes_organtype[:,3] ==1 )] # remove seed. get roots with one connection

        organ_number_to_order = [0] *len(organ)
        for k in reversed(range(len(organ))):  # loop through all the organs
            if organ[k].getParent().organType() == 1:
                organ_number_to_order[k] = 1
            elif organ[k].getParent().getParent().organType() == 1:
                organ_number_to_order[k] = 2
            elif organ[k].getParent().getParent().getParent().organType() == 1:
                organ_number_to_order[k] = 3
            elif organ[k].getParent().getParent().getParent().getParent().organType() == 1:
                organ_number_to_order[k] = 4
            elif organ[k].getParent().getParent().getParent().getParent().getParent().organType() == 1:
                organ_number_to_order[k] = 5
            else:
                organ_number_to_order[k] = 6

        i=0
        for i in range(len(nodes_organtype)):
            oder_number = organ_number_to_order[node_id_to_organ[i]]
            pd_output_1day.loc[i,"order"] = oder_number
            pd_output_1day.loc[i, "radius"] = plant.getOrgans()[node_id_to_organ[i]].getParameter("a")
            if oder_number == 1:
                pd_output_1day.loc[i,"parent_organ"] = 0
            else:
                pd_output_1day.loc[i,"parent_organ"] = organ[node_id_to_organ[i]].getParent().getId()

        radial_r_xylem =  []     
        axial_r_xylem = []
        
        for i in range(1,(len(dict_all["nodes_organtype"]))):
            radius = plant.getOrgans()[node_id_to_organ[i]].getParameter("a")
            connection_belong = np.where(node_connection_o.T[1][:] ==i )
            first_node = node_connection_o[connection_belong[0][0]][0]
            node_length= nodes_length[i] - nodes_length[first_node]
            surface = abs(node_length)*100 * 3.1415926 * radius**2
            if organ_number_to_order[node_id_to_organ[i]] == 1:
                radial_r_xylem.append( 
                    ky_to_r(r_basal.predict(max(0 , organ[node_id_to_organ[i]].getLength() - nodes_length[i]*100)))[0] /surface  )      
            else:

                radial_r_xylem.append( ky_to_r(r_lateral.predict(day_i - plant.getNodeCTs()[i]))[0] /surface  )
        radial_r_xylem.insert(0,radial_r_xylem[0] )
#         print(radial_r_xylem)
        
        for i in range(0,(len(dict_all["node_connection"]))):
            second_node = dict_all["node_connection"].T[1][i]-1 # the node connection starts with node 1
            first_node = dict_all["node_connection"].T[0][i]-1
            if plant.getOrgans()[node_id_to_organ[second_node]].getParameter("organType") == 2:
                if organ_number_to_order[node_id_to_organ[second_node]] == 1:
                    axial_r_xylem.append(
                       bk_to_r(a_basal.predict(max(0 ,60)))[0] *0.25/1000 # in doussan it was /cm here is per 0.25 cm 
                    
                    )

                else:
                    axial_r_xylem.append( lk_to_r(a_lateral.predict(day_i - plant.getSegmentCTs()[i]))[0] *0.25/1000  )
            else:
                axial_r_xylem.append( 1e-9 )
                
        print(axial_r_xylem)

        

        axial_r_phloem = []

        for i in range(0,(len(dict_all["node_connection"]))):

            phloem_ratio =420
            
            second_node = dict_all["node_connection"].T[1][i]-1 # the node connection starts with node 1
            first_node = dict_all["node_connection"].T[0][i]-1 # the node connection starts with node 0
            if plant.getOrgans()[node_id_to_organ[second_node]].getParameter("organType") == 2:
                if plant.getOrgans()[node_id_to_organ[second_node]].getParameter("subType") == 1:
                    phloem_radius = plant.getOrgans()[node_id_to_organ[second_node]].getParameter("a")/phloem_ratio
                    resistance_propotional =8 * 1.7 *(abs(nodes_length[first_node] - 
                                nodes_length[second_node]) *100 )/(3.1415926* phloem_radius**4) / 10e9/3600 l
                    axial_r_phloem.append(resistance_propotional*10/5000 *(0.5*2** (int(resistance_parameter))))
                else:
                    phloem_radius = plant.getOrgans()[node_id_to_organ[second_node]].getParameter("a")/phloem_ratio
                    resistance_propotional =8 * 1.7 *(abs(nodes_length[first_node] - 
                                            nodes_length[second_node]) *100 )/(3.1415926* phloem_radius**4) / 10e9/3600 
                    axial_r_phloem.append(resistance_propotional*10/5000*3)
            else:
                axial_r_phloem.append(1e-4)
        ################################# coupling layer with piafmunch ###################################################
        homo_st = [0.04]*(len(dict_all["node_connection"]))
        nodes_organtype =dict_all['nodes_organtype']
        node_connection = dict_all['node_connection']
        unq_cnt = nodes_organtype.T[3]
        write_PiafMunch_parameter(node_connection, 
                                  nodes_organtype, 
                                   axial_r_phloem, 
                                  unq_cnt, 
                                  Soil_water, 
                                  k1, 
                                  piafmunch_outputname,
                                  hour, 
                                  axial_r_xylem, 
                                  vml, 
                                  vol_st, 
                                  vmu, 
                                  r_phl_mb,
                                  r_trsv)
        # call piafmunch and rename the output file of the PiafMunch
        os.system('cp {}  ./{}_{}_{}_{}_{}_{}_input.ini'.format(piafmunch_outputname,name, time_total, resistance_parameter, loading_vml, unloading_vmu,day_i ) )
        os.system(' /mnt/c/prog_sct.bas/programs/C++/PIAFMunch2/projects/PIAFMunch2.exe {} -noGUI -q'.format(piafmunch_outputname))
        now = datetime.datetime.now()
        os.system('mv ./{}{}_{}-{:02}-{:02}*output.txt  ./{}_{}_{}_{}_{}_{}_output.txt'.format(name,day_i, now.year ,now.month, now.day-1,name, time_total, resistance_parameter, loading_vml, unloading_vmu,day_i ) )
        os.system('mv ./{}{}_{}-{:02}-{:02}*output.txt  ./{}_{}_{}_{}_{}_{}_output.txt'.format(name,day_i, now.year ,now.month, now.day,name, time_total, resistance_parameter, loading_vml, unloading_vmu,day_i ) )
        output_name = "{}_{}_{}_{}_{}_{}".format(name, time_total, resistance_parameter, loading_vml, unloading_vmu, day_i)
        ########################## convert carbon to growth rate ###########################################
        piaf_output = read_output(output_name,  dict_all["node_connection"])
        for node_i,root_node in enumerate(N1R_node): # CPlantBox growth rate: cm day-1
            nodes_organ = organ[node_id_to_organ[int(root_node[0])-1]]
            pd_output_1day.loc[int(root_node[0])-1,"organ_length"] = nodes_organ.getLength()
            cross_area_root= nodes_organ.getParam().a**2*3.1415926 # this is also the carbon used for growth because mass need is 1 g DM cm3
            connection_belong = np.where(node_connection_o.T[1][:] ==int(root_node[0])-1 )
            carbon_available = piaf_output["JS_ST"][-3][connection_belong[0][0]]
            carbon_available_g_c = carbon_available*12* 12/1000 *24  #  # gram of carbon per day
            rw_satis = min(1,max(0, g_propotion(piaf_output["P_Xyl"][-3][int(root_node[0]-1)])))
            groth_rate = subtype_to_growth_rate[int(nodes_organ.getParam().subType)]   
            gr_water = rw_satis *groth_rate  # rw_satis * 8
            gr_pot_2 = gr_water
            c_demand_2 = max(1e-9, cross_area_root *gr_pot_2  * 0.1 *0.4 ) # 0.1 g DM cm3 / day
            c_offer_2 = carbon_available_g_c
            c_satis = c_offer_2 / c_demand_2
            gr_act = gr_pot_2 * max([0.000001, min([1,c_satis])])
            nodes_organ.getParam().r = gr_act
            pd_output_1day.loc[int(root_node[0])-1, "growth_rate"] = gr_act
            pd_output_1day.loc[int(root_node[0])-1, "exudation"] = max([0 ,carbon_available_g_c - c_demand_2])  
            pd_output_1day.loc[int(root_node[0])-1,"carbon_q"] = carbon_available_g_c
            pd_output_1day.loc[int(root_node[0])-1,"connection_belong"] = connection_belong[0][0]
            
            pd_output_1day.loc[int(root_node[0])-1, "rw_satis"] = rw_satis
            pd_output_1day.loc[int(root_node[0])-1, "c_satis"] = c_satis
            pd_output_1day.loc[int(root_node[0])-1, "c_demand_2"] = c_demand_2


        ############################################ log ###########################
        pd_output_1day.loc[:, "JW_Trsv"] =  piaf_output["JW_Trsv"][-3]
        pd_output_1day.loc[:, "Q_ST"] =  piaf_output["Q_ST"][-3]
        pd_output_1day.loc[:, "C_ST"] =  piaf_output["C_ST"][-3]
        
        pd_output_1day.loc[:, "P_Xyl"] =  piaf_output["P_Xyl"][-3]
        pd_output_1day.loc[:, "P_ST"] =  piaf_output["P_ST"][-3]
        pd_output_1day.loc[:,"organ_id"] = list(map(int,node_id_to_organ))
        pd_output_1day.loc[:,"organ_id2"] = list(map(int,node_id_to_organ2))

        pd_output_1day.loc[:,"length"] = list(nodes_length)
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"r_xyl"] = list(axial_r_xylem)
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"r_st"] = list(axial_r_phloem)
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"JW_Xyl"] = piaf_output["JW_Xyl"][-3]
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"JW_ST"] = piaf_output["JW_ST"][-3]
        pd_output_1day.loc[:len(dict_all["node_connection"])-1,"JS_ST"] = piaf_output["JS_ST"][-3]
        
        pd_output_1day.loc[:,"res_main"] = list(k1)
        
        pd_output_1day.loc[:,"time"] = day_i
        pd_output_1day.loc[:,"sub_type"] = list(node_id_to_subtype)
        pd_output_1day.loc[0,"sub_type"] = 1
        pd_output = pd_output.append(pd_output_1day)


    pd_output.to_csv("{}-{}-{}-{}-{}.csv".format(output_name,now.month,now.day,now.hour,now.minute), index=False)# %%time


```
